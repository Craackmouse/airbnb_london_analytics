## 2.Price Heatmap:

Use the calendar dataset to extract average prices of listings in each neighbourhood for a given time frame.

Plot a heatmap on the map of London to identify high-priced and low-priced neighbourhoods.

```python
# Convert the price column in calendar_df to a numerical format
# calendar_df['price'] = calendar_df['price'].str.replace(',', '').str.replace('$', '')

# Now you can convert the 'price' column to a numerical type (e.g., float)
calendar_df['price'] = calendar_df['price'].astype(float)
# Merge the listings and calendar datasets to get the price for each listing
merged_df = pd.merge(listings_plain_df[['id', 'neighbourhood']], calendar_df[['listing_id', 'price']], 
                     left_on='id', right_on='listing_id')

# Group by neighbourhood and compute the average price
avg_price_by_neighbourhood = merged_df.groupby('neighbourhood')['price'].mean().reset_index()

# Merge the aggregated data with the neighbourhoods_geo dataset for visualization
merged_geo_df = pd.merge(neighbourhoods_geo_df, avg_price_by_neighbourhood, on='neighbourhood', how='left')

# Visualize the heatmap based on average price
fig, ax = plt.subplots(figsize=(15, 15))
merged_geo_df.plot(column='price', cmap='OrRd', linewidth=0.8, edgecolor='0.8', 
                   legend=True, ax=ax, legend_kwds={'label': "Average Price (£)"})
ax.set_title("Average Airbnb Listing Price by Neighbourhood in London")
plt.axis('off')
plt.tight_layout()
plt.show()
```

Observation:

Central London: As expected, the central parts of London tend to have highest average prices for Airbnb listings.

Peripheral Areas: Some of the northerna and western peripheral areas have higher average prices, which can be due to airport and other attractions, or transportation hubs.

## 3.Availability Analysis:

Use the calendar dataset to determine the availability of properties in each neighbourhood.

Visualize areas with high availability (potential low demand) and low availability (potential high demand).

```python
# Convert the available column in calendar_df to a numerical format (1 for 't' and 0 for 'f')
calendar_df['availability'] = calendar_df['available'].map({'t': 1, 'f': 0})

# Merge the listings and calendar datasets to get the availability for each listing
availability_df = pd.merge(listings_plain_df[['id', 'neighbourhood']], calendar_df[['listing_id', 'availability']], 
                           left_on='id', right_on='listing_id')

# Group by neighbourhood and compute the average availability
avg_availability_by_neighbourhood = availability_df.groupby('neighbourhood')['availability'].mean().reset_index()

# Merge the aggregated data with the neighbourhoods_geo dataset for visualization
availability_geo_df = pd.merge(neighbourhoods_geo_df, avg_availability_by_neighbourhood, on='neighbourhood', how='left')

# Visualize the heatmap based on average availability
fig, ax = plt.subplots(figsize=(15, 15))
availability_geo_df.plot(column='availability', cmap='RdYlGn_r', linewidth=0.8, edgecolor='0.8', 
                         legend=True, ax=ax, legend_kwds={'label': "Average Availability"})
ax.set_title("Average Airbnb Listing Availability by Neighbourhood in London")
plt.axis('off')
plt.tight_layout()
plt.show()
```

## 4.Review rate 
Count the number of reviews for each listing to estimate the number of times it was occupied

Calculate the occupancy rate as:

Review Rate = Number of Reviews/Total Days,

where Total Days is the time duration for which we have the reviews data.

```python
# Count the number of reviews for each listing
reviews_count_reload = reviews_df.groupby('listing_id').size().reset_index(name='num_reviews')

# Calculate the total number of days for which we have reviews data
reviews_df['date'] = pd.to_datetime(reviews_df['date'])
total_days_reload = (reviews_df['date'].max() - reviews_df['date'].min()).days

# Calculate the occupancy rate
reviews_count_reload['occupancy_rate'] = reviews_count_reload['num_reviews'] / total_days_reload

# Merge this data with the listings dataset
listings_with_occupancy_reload = pd.merge(listings_df,
                                          reviews_count_reload[['listing_id', 'occupancy_rate']],
                                          left_on='id', right_on='listing_id', how='left')

# Fill NaN values with 0 (for listings with no reviews)
listings_with_occupancy_reload['occupancy_rate'].fillna(0, inplace=True)

# Convert listings data to a GeoDataFrame using latitude and longitude
listings_with_occupancy_gdf_reload = gpd.GeoDataFrame(listings_with_occupancy_reload, 
                                                      geometry=gpd.points_from_xy(listings_with_occupancy_reload.longitude, listings_with_occupancy_reload.latitude),
                                                      crs="EPSG:4326")

# Plot the map with occupancy rates
fig, ax = plt.subplots(figsize=(15, 15))
neighbourhoods_geo_df.plot(ax=ax, color='green', edgecolor='black')
listings_with_occupancy_gdf_reload.plot(column='occupancy_rate', cmap='OrRd', legend=True, ax=ax, markersize=2, 
                                        legend_kwds={'label': "Occupancy Rate"})
ax.set_title("Occupancy Rate of Airbnb Listings in London")
plt.axis('off')
plt.tight_layout()
plt.show()
```

## 5.1.Neighbourhood Popularity based on review:

Use the number of reviews to determine the popularity of each neighbourhood.

Highlight popular neighbourhoods on the map.

```python
# Count the number of reviews for each listing
review_counts = reviews_df['listing_id'].value_counts().reset_index()
review_counts.columns = ['listing_id', 'review_count']

# Merge the review counts with the listings dataset
listings_with_reviews = pd.merge(listings_plain_df, review_counts, left_on='id', right_on='listing_id', how='left')

# Fill NA values with 0 for listings that have no reviews
listings_with_reviews['review_count'].fillna(0, inplace=True)

# Aggregate the review counts by neighbourhood
neighbourhood_popularity = listings_with_reviews.groupby('neighbourhood')['review_count'].sum().reset_index()

# Merge the aggregated data with the neighbourhoods_geo dataset for visualization
neighbourhoods_with_popularity = pd.merge(neighbourhoods_geo_df, neighbourhood_popularity, 
                                         on='neighbourhood', how='left')

# Visualize the heatmap based on neighbourhood popularity (review count)
fig, ax = plt.subplots(figsize=(15, 15))
neighbourhoods_with_popularity.plot(column='review_count', cmap='OrRd', linewidth=0.8, edgecolor='0.8', 
                                    legend=True, ax=ax, legend_kwds={'label': "Review Count"})
ax.set_title("Popularity of Neighbourhoods in London (Based on Review Count)")
plt.axis('off')
plt.tight_layout()
plt.show()
```

Observation:
    
Central and well-known areas of London would typically have higher review counts, indicating that they are popular tourist destinations.

Price and popularity heatmap has overlap as expected

## 5.2.Neighbourhood Popularity based on frequency of booking:

Use the number of booking frequency to determine the popularity of each neighbourhood.

Highlight popular neighbourhoods on the map.

```python
booked_dates = calendar_df[calendar_df['available'] == 'f']
booking_counts_per_listing = booked_dates['listing_id'].value_counts().reset_index()
booking_counts_per_listing.columns = ['listing_id', 'booking_count']
merged_df = pd.merge(listings_plain_df, booking_counts_per_listing, left_on='id', right_on='listing_id', how='left')
merged_df['booking_count'].fillna(0, inplace=True)
neighbourhood_booking_counts = merged_df.groupby('neighbourhood')['booking_count'].sum().reset_index()
merged_geo_booking_df = pd.merge(neighbourhoods_geo_df, neighbourhood_booking_counts, on='neighbourhood', how='left')

# Visualization based on booking frequency
fig, ax = plt.subplots(figsize=(15, 15))
merged_geo_booking_df.plot(column='booking_count', ax=ax, edgecolor='black', cmap='OrRd', legend=True)
ax.set_title("Popularity of Neighbourhoods in London based on Booking Frequency")
plt.axis('off')
plt.tight_layout()
plt.show()
```

## 7. Price vs availibility
Based on calander visualise price to availibility analysis


```python
# Convert the 'date' column to datetime format for sorting and plotting purposes
calendar_df['date'] = pd.to_datetime(calendar_df['date'])

# Calculate the average price for each date
average_price_by_date = calendar_df.groupby('date')['price'].mean()

# Calculate the percentage of listings that are available for each date
availability_by_date = calendar_df.groupby('date')['available'].apply(lambda x: (x == 't').sum() / len(x) * 100)

# Plotting
fig, ax1 = plt.subplots(figsize=(15, 7))

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Average Price (£)', color=color)
ax1.plot(average_price_by_date.index, average_price_by_date.values, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Percentage of Listings Available (%)', color=color)
ax2.plot(availability_by_date.index, availability_by_date.values, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('Temporal Patterns: Average Listing Price and Availability Over Time')
plt.show()
```

Observation:

Availibility increasing as we look further where as price decreasing

Availibility dip during new year season and price shot on new year night


## 8.Expected(future) Temporal Trends by listings:

Analyze how prices or availability change over time (e.g., seasonally) for each neighbourhood based on calander future data.

Visualize areas with the highest price hikes during peak seasons or events.

```python
# Convert the 'date' column to datetime format
calendar_df['date'] = pd.to_datetime(calendar_df['date'])

# Convert the 'price' column to numeric (removing any non-numeric characters if present)
calendar_df['price'] = calendar_df['price'].replace('[\$,]', '', regex=True).astype(float)

# Merge with listings to get neighbourhood for each listing
calendar_with_neighbourhood = calendar_df.merge(listings_plain_df[['id', 'neighbourhood']], left_on='listing_id', right_on='id', how='left')

# Calculate the average price for each date for each neighbourhood
average_price_by_date_neighbourhood = calendar_with_neighbourhood.groupby(['date', 'neighbourhood'])['price'].mean().reset_index()

average_price_by_date_neighbourhood.head()
# Selecting a subset of major neighbourhoods for clearer visualization
selected_neighbourhoods = ["Westminster", "Camden", "Kensington and Chelsea", "Islington", "Tower Hamlets", "Hackney"]

filtered_avg_price = average_price_by_date_neighbourhood[average_price_by_date_neighbourhood['neighbourhood'].isin(selected_neighbourhoods)]

# Plotting the trends for expected price changes
plt.figure(figsize=(16, 8))
for neighbourhood in selected_neighbourhoods:
    subset = filtered_avg_price[filtered_avg_price['neighbourhood'] == neighbourhood]
    plt.plot(subset['date'], subset['price'], label=neighbourhood)

plt.title("Expected Monthly Average Pricing Trends by Neighbourhood")
plt.xlabel("Date")
plt.ylabel("Average Expected Price (£)")
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
